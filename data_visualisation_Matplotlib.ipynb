```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=[2,3,4,5,6,7]
y=[10,20,30,40,50,60]
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x=[2,3,4,5,6,7]
y=[10,20,30,40,50,60]
a=[100,80,60,70,50,40]
plt.plot(x,y, color='green',linestyle=':',label="vs y",marker="s")
plt.plot(x,a, color='blue',linestyle='--',label="vs y",marker="^")
plt.title("Basic Plotting using Matplotlib")
plt.xlabel("Height")
plt.ylabel("Weight/Volume")
plt.legend()
plt.grid(False)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
fig,ax=plt.subplots(2,1,figsize=(6,6))
ax[0].plot(x,y, color="green", linestyle="--",label="vs Y",marker='s')
ax[0].set_xlabel("Height")
ax[0].set_ylabel("Weight")
ax[0].legend(loc=9)
ax[0].set_title("Weight vs Height Graph - 1")
ax[1].plot(x,a, color="blue", linestyle=":",label="vs Y",marker='^')
ax[1].set_xlabel("Height")
ax[1].set_ylabel("Weight")
ax[1].legend()
ax[1].set_title("Weight vs Height Graph - 2")
fig.suptitle("Plot Super Title",fontsize=20)
fig.tight_layout()
plt.show()
```

```python
import seaborn as sns
tips = sns.load_dataset('tips') # Sample Dataset : for the examples
tips.head()
```

```python
sns.histplot(x=tips['total_bill'],color='green',kde=True,bins=10)
plt.title("Distribution Of Total Bill")
plt.xlabel("Amount Of total Bill")
plt.ylabel("Count Of Bill")
plt.show()
```

```python
sns.histplot(data=tips,x='total_bill',color='green',kde=True,bins=10,hue='time')
plt.title("Distribution Of Total Bill")
plt.xlabel("Amount Of total Bill")
plt.ylabel("Count Of Bill")
plt.show()
```

```python
tips['time'].value_counts()
```

```python
sns.barplot(x='day',y='total_bill',data=tips,color='green')
plt.title("Average Total bill by day")
plt.xlabel("Week days")
plt.ylabel("Avg total bill")
plt.show()
```

```python
sns.countplot(x='day',data=tips,color='green')
```

```python
sns.boxplot(x='day',y=tips['tip'],color='green',data=tips,hue='smoker')
```

```python
tips[tips['total_bill']<16]
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,color='red')
```

```python
tips_grouped=tips.groupby('size').agg({'tip':'mean','total_bill':'mean'}).reset_index()
```

```python
tips_grouped
```

```python
sns.lineplot(data=tips_grouped,x='size',y='total_bill',color="green",marker='*')
```

```python
fig,ax = plt.subplots(2,2,figsize=(10,10))
# Subplot-1
sns.histplot(data=tips,x='total_bill',ax=ax[0,0],color='green',kde=True,hue='smoker')
ax[0,0].set_title("Distribution Of Total Bill",fontdict={'fontsize':10,'fontweight':'bold'})
ax[0,0].set_xlabel('Total Bill',fontdict={'fontsize':10,'fontweight':'bold'})
ax[0,0].set_ylabel('Count Of Bill',fontdict={'fontsize':10,'fontweight':'bold'})
ax[0,0].tick_params(axis='x',rotation=45,labelsize=10)


#  Subplot 2 : Barplot
sns.barplot(x='day', y='total_bill', data=tips, ax = ax[0,1],color = 'green',hue = 'time')
# If no preparation on raw data --> Default (group by [x and Hue] --> mean of y)
ax[0,1].set_title('Total Bill Across Day', fontdict={'fontsize': 10, 'fontweight': 'bold'})
ax[0,1].set_xlabel('Weekday', fontdict={'fontsize': 8, 'fontweight': 'bold'})
ax[0,1].set_ylabel('Avg total Bill', fontdict={'fontsize': 8, 'fontweight': 'bold'})
ax[0,1].tick_params(axis='x', labelsize=6, rotation=45)
ax[0,1].tick_params(axis='y', labelsize=6)
ax[0,1].axhline(y= tips['total_bill'].median(), color='red', linestyle='--', linewidth=2, label='Median Total Bill')
ax[0,1].text(0.25, 19, '** Global Median Value', fontsize=6, color='red')

# horizontal line --> on which days avg goes beyond the overall average
# Subplot 3:
sns.boxplot(x='day', y=tips['tip'], data=tips, ax = ax[1,0])
ax[1,0].set_title('Box Plot of Total Bill by Day', fontdict={'fontsize': 10, 'fontweight': 'bold'})
ax[1,0].set_xlabel('Day', fontdict={'fontsize': 8, 'fontweight': 'bold'})
ax[1,0].set_ylabel('Total Bill', fontdict={'fontsize': 8, 'fontweight': 'bold'})
ax[1,0].text(2, 10, '** Notice Outliers', fontsize=6, color='red')
ax[1,0].set_xticks([0,1,2,3]) # Position of labels
ax[1,0].set_xticklabels(labels=['Thursday', 'Friday', 'Saturday', 'Sunday'],fontsize= 6, rotation = 45)

# Subplot 4:
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', ax = ax[1,1])
ax[1,1].set_title('Scatter Plot of Total Bill vs. Tip', fontdict={'fontsize': 10, 'fontweight': 'bold'})
ax[1,1].set_xlabel('Total Bill', fontdict={'fontsize': 8, 'fontweight': 'bold'})
ax[1,1].set_ylabel('Tip', fontdict={'fontsize': 8, 'fontweight': 'bold'})
ax[1,1].tick_params(axis='x', labelsize=6, rotation=45)
ax[1,1].tick_params(axis='y', labelsize=6)


fig.savefig('/content/drive/MyDrive/M2-Class/4plotnew.png',dpi=300)
fig.suptitle("Analysis of Tips Dataset", fontsize=20)
fig.tight_layout()
plt.show()
```
